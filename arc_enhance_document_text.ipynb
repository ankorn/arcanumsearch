```python
from datasets import load_dataset

dataset = load_dataset("pameydorke/arcanum-quests-queries-synthetic")
```

```python
REMOVE_SUBSTRINGS = [': Trivia', ': Walkthrough', ': Notes', 'Path 1, ', 'Path 2, ', 'Path 3, ', ': Short Description', ': Background', 'Part 01 - ', 'Part 02 - ', 'Part 03 - ', 'Part 04 - ', 'Part 05 - ', 'Part 06 - ', 'Part 07 - ', 'Part 08 - ', 'Part 09 - ', 'Part 10 - ', 'Part 11 - ', 'Part 12 - ', 'Part 13 ', 'Part 13a ', 'Part 13b ', 'Part 13c ', 'Part 14 - ', ': Epilogue', 'optional continuation']
TITLE_MAP = {
      "Find the Bow of Ecclesiastes": "Skill mastery quest. Bow mastery. " + "Find the Bow of Ecclesiastes",
      "Kill Sir Garrick Stout": "Skill mastery quest. Dodge mastery. " + "Kill Sir Garrick Stout",
      "Find Lady Druella": "Skill mastery quest. Melee mastery. " + "Find Lady Druella",
      "Retrieve Azram's Star": "Skill mastery quest. Throwing mastery. " + "Retrieve Azram's Star",
      "Find the Master of Backstab": "Skill mastery quest. Backstab mastery. " + "Find the Master of Backstab",
      "Run Around Tarant in Your Underwear": "Skill mastery quest. Pick pocket mastery. " + "Run Around Tarant in Your Underwear",
      "Find the Master of Prowling": "Skill mastery quest. Prowling mastery. " + "Find the Master of Prowling",
      "Get staff of K’an T’au": "Skill mastery quest. Spot trap mastery. " + "Get staff of K’an T’au",
      "Gamble with Gurin Rockharrow": "Skill mastery quest. Gambling mastery. " + "Gamble with Gurin Rockharrow",
      "Acquire Ten Thousand Gold Pieces": "Skill mastery quest. Haggle mastery. " + "Acquire Ten Thousand Gold Pieces",
      "Find the Master of Heal": "Skill mastery quest. Heal mastery. " + "Find the Master of Heal",
      "Negotiations with Caladon": "Skill mastery quest. Persuasion mastery. " + "Negotiations with Caladon",
      "Find Proof that Maxim's Air Machines Flew": "Skill mastery quest. Repair mastery. " + "Find Proof that Maxim's Air Machines Flew",
      "Rescue Mrs. Rolland Unharmed": "Skill mastery quest. Firearms mastery. Guns mastery. " + "Rescue Mrs. Rolland Unharmed",
      "Free J.T. Morgan": "Skill mastery quest. Pick locks mastery. Steal mastery. " + "Free J.T. Morgan",
      "Survive the Training Maze": "Skill mastery quest. Disarm traps mastery. " + "Survive the Training Maze",
    }
```

```python
def enhance_title(title: str) -> str:
    cleaned = title
    for sub in REMOVE_SUBSTRINGS:
        cleaned = cleaned.replace(sub, '')

    # 'Find the Bow of Ecclesiastes: Walkthrough' -> 'Find the Bow of Ecclesiastes. Walkthrough. '
    parts = cleaned.split(': ')
    formatted = '. '.join(parts)
    if not formatted.endswith('. '):
        formatted += '. '

    base_title = formatted.rstrip('. ')
    if base_title in TITLE_MAP:
        formatted = TITLE_MAP[base_title] + '. '

    return formatted

def process_example(example):
    formatted_title = enhance_title(example['document_title'])
    example['enhanced_document_text'] = formatted_title + example['document_text']
    return example
```

```python
for split in dataset.keys():
    dataset[split] = dataset[split].map(process_example)
```
